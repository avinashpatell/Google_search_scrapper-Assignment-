```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import csv
import time
```

```python
categories1 = [
    "Manufacturer", "Brand", "Distributor", "F&B", "Probiotics",
    "Fortification", "Gut Health", "Womens Health", "Cognitive Health",
    "Research & Development"
]

companiesL = [
       [
        "Nestle", "Dr. Reddy's Laboratories", "Coca", "Pfizer", "PepsiCo",
        "Johnson & Johnson", "Danone", "Bayer", "General Mills",
        "GlaxoSmithKline (GSK)", "Kellogg’s", "Merck & Co.", "Unilever", "Roche"
    ],
    [
        "Nestle Waters", "Sanofi", "Mondelez International", "Novartis", "Kraft Heinz",
        "Eli Lilly and Company", "Tyson Foods", "Teva Pharmaceuticals", "Mars, Incorporated",
        "AbbVie", "Campbell Soup Company", "Amgen", "Conagra Brands", "AstraZeneca"
    ],
    [
        "Molson Coors", "Boehringer Ingelheim", "AB InBev", "BASF", "Diageo",
        "Procter & Gamble (P&G)", "Heineken", "Medtronic", "McKesson", "AmerisourceBergen",
        "Cardinal Health", "Medline Industries"
    ]
]
i = 1
for companies in companiesL:
    options = uc.ChromeOptions()
    options.headless = False
    driver = uc.Chrome(use_subprocess=False, options=options)
    
    driver.get("https://www.google.com")
    time.sleep(2)
    
    csv_file = f"output{i}.csv"
    with open(csv_file, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file)
    
        # header
        writer.writerow(["Company Name"] + categories1)
    
        for company in companies:
            row_data = [company]  # company name
            print(f"{company} \n {categories1} \n")
    
            for category in categories1:
                negatives = ["No", "no", "does not", "doesn't", "not"]
                ques = f"Is {company} a {category} company?"
                answer = "No"
    
                try:
                    input_element = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.NAME, "q"))
                    )
    
                    input_element.clear()
                    input_element.send_keys(ques)
                    input_element.send_keys(Keys.ENTER)
    
                    time.sleep(3)
    

                    try:
                        checkstr = driver.find_element(By.CLASS_NAME, "rPeykc")
                        if checkstr.text not in negatives and ("Yes" in checkstr.text or "yes" in checkstr.text):
                           answer = "Yes"
                        else:
                            answer = "No"
                    except NoSuchElementException:
                        print(f"Answer not found for: {ques}")
    
                except TimeoutException:
                    print(f"Timeout for: {ques}")
    
                row_data.append(answer)
                print(row_data)
                time.sleep(2)
                driver.get("https://www.google.com") 
    
            writer.writerow(row_data)
            print(f"Saved: {company} -> {row_data[1:]}")
    
    i +=1
    driver.quit()
    time.sleep(100)
    
```
